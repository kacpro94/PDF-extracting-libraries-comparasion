## PDFplumber

```python
tables_all=[]
try:
    with pdfplumber.open("data/invoices/wordpress-pdf-invoice-plugin-sample.pdf") as pdf:
        for i, page in enumerate(pdf.pages):
            tables = page.extract_tables()

            for table_data in tables:
                    if not table_data or len(table_data) < 1:
                        continue
                    headers = table_data[0]
                    rows = table_data[1:]
                    
                    df = pd.DataFrame(rows)
                    df.columns = headers
                    tables_all.append(df)
except Exception as e:
        print(f"❌ Błąd podczas przetwarzania: {e}")

for table in tables_all:
     display(table)
```

Wynik uzyskany przy pomocy pdfplumber jest identyczny jak przy PyMuPDF. Jest to zapewne związane z tym, ze obydwie bibioteki wchodza w kod zrodlowy pdf i na podstawie tego okreslaja gdzie jest tabela itp. 

## Docling

```python
results = []
print("Testowanie Docling...")
start = time.perf_counter()
text_docling = ""
t_docling = None
score_docling = None
docling_dfs = []

try:
    converter = DocumentConverter()
    res = converter.convert("data/invoices/wordpress-pdf-invoice-plugin-sample.pdf")
    try:
        md = res.document.export_to_markdown()
    except Exception:
        md = ""
        for page in res.document.pages:
            txt = page.export_to_markdown()
            if txt:
                md += txt + "\n"
    text_docling = md
    for table in res.document.tables:
        df = table.export_to_dataframe()
        docling_dfs.append(df)
    t_docling = time.perf_counter() - start
    
except Exception as e:
    print(f"Docling Error: {e}")
    t_docling = time.perf_counter() - start
    

print(f"time:{t_docling}")
print("\n")
print(text_docling)
for i, df in enumerate(docling_dfs):
    print(f"\n--- Tabela #{i+1} (Docling) ---")
    display(df)
```

Efekty są zdecydowanie najlepsze. Jak widać, tej bibiotece udało się w 100% poprawnie wyciągnąc 3 tabele i nawet samemu ustalic nazwy kolumn. Zdecydowanie dluzej trwal ten proces, ale efekty są w zupelnosci poprawne.
