```python
def benchmark_libraries(pdf_path, ground_truth):
    results = []
    print("Testowanie PyMuPDF...")
    start = time.perf_counter()
    try:
        doc = fitz.open(pdf_path)
        text_pymu = ""
        for page in doc:
            text_pymu += page.get_text() + "\n"
        t_pymu = time.perf_counter() - start
        score_pymu = fuzz.ratio(text_pymu.strip(), ground_truth.strip())
        results.append({"Lib": "PyMuPDF", "Time": t_pymu, "Score": score_pymu, "Text": text_pymu})
    except Exception as e:
        print(f"PyMuPDF Error: {e}")

    print("Testowanie pdfplumber...")
    start = time.perf_counter()
    try:
        text_plumb = ""
        with pdfplumber.open(pdf_path) as pdf:
            for page in pdf.pages:
                text_plumb += (page.extract_text() or "") + "\n"
        t_plumb = time.perf_counter() - start
        score_plumb = fuzz.ratio(text_plumb.strip(), ground_truth.strip())
        results.append({"Lib": "pdfplumber", "Time": t_plumb, "Score": score_plumb, "Text": text_plumb})
    except Exception as e:
        print(f"pdfplumber Error: {e}")

    print("Testowanie Docling (AI)...")
    start = time.perf_counter()
    try:
        converter = DocumentConverter()
        result = converter.convert(pdf_path)
        text = result.document.export_to_markdown()
        
        duration = time.perf_counter() - start
        score = fuzz.ratio(text.strip(), ground_truth.strip())
        results.append({"Lib": "Docling (AI)", "Time": duration, "Score": score, "Text": text})
    except Exception as e:
        print(f"   B≈ÇƒÖd: {e}")

    print("PdfPlumber"+"="*40)
    print(text_plumb)
    print("PyMuPDF"+"="*40)
    print(text_pymu)
    print("Docling"+"="*40)
    print(text)

    print("\n" + "="*40)
    print(f"{'BIBLIOTEKA':<15} | {'CZAS (s)':<10} | {'DOK≈ÅADNO≈öƒÜ':<10}")
    print("-" * 40)
    for res in results:
        print(f"{res['Lib']:<15} | {res['Time']:.4f}     | {res['Score']:.2f}%")
    print("="*40)
    
    return results

wyniki = benchmark_libraries("data/texts/oldmansea-20.pdf", ground_truth_simple_1)
```

```python
import pandas as pd
import pdfplumber
from docling.document_converter import DocumentConverter
import re

# ≈öcie≈ºka do folderu
PDF_PATH = "data/invoices"

# 1. GROUND TRUTH (Wzorcowe dane)
GT_INVOICE_NUM = "INV-3337"
GT_RECIPIENT = "Test Business"

# 2. FUNKCJE POMOCNICZE
def clean_text(text):
    if not isinstance(text, str): return str(text)
    text = text.replace("\n", " ").replace("\r", "")
    return re.sub(' +', ' ', text).strip()

# Globalny s≈Çownik na wyniki
final_report = {}

def evaluate_extraction(lib_name, invoice_num, recipient, extracted_df):
    """Ocenia jako≈õƒá ekstrakcji w spos√≥b ≈õcis≈Çy (Strict Mode)."""
    print(f"\n--- üïµÔ∏è‚Äç‚ôÇÔ∏è OCENA DLA: {lib_name} ---")
    
    # --- 1. Pola tekstowe ---
    inv_status = "‚úÖ OK" if invoice_num and GT_INVOICE_NUM in invoice_num else f"‚ùå B≈ÅƒÑD (Oczek: {GT_INVOICE_NUM})"
    rec_status = "‚úÖ OK" if recipient and GT_RECIPIENT in recipient else f"‚ùå B≈ÅƒÑD (Oczek: {GT_RECIPIENT})"
    
    print(f"Numer faktury: {invoice_num} -> {inv_status}")
    print(f"Odbiorca:      {recipient} -> {rec_status}")
    
    # --- 2. Tabela (Analiza Strukturalna) ---
    table_status = "‚ùå Tabela nie znaleziona"
    details = []
    
    if extracted_df is not None and not extracted_df.empty:
        # Normalizacja
        extracted_df = extracted_df.map(clean_text)
        extracted_df.columns = [clean_text(col) for col in extracted_df.columns]
        
        rows_count = len(extracted_df)
        cols_count = len(extracted_df.columns)
        
        # DEFINICJA WARTO≈öCI KLUCZOWYCH (MUSZƒÑ BYƒÜ W JEDNYM WIERSZU)
        key_price = "$85.00"
        key_qty = "1.00"
        key_desc_part = "Web Design"
        
        # Krok A: Szukamy "Idealnego Wiersza"
        found_perfect_row = False
        found_fragmented_data = False
        
        for idx, row in extracted_df.iterrows():
            row_str = " ".join(row.values.astype(str))
            
            has_price = key_price in row_str
            has_qty = key_qty in row_str
            has_desc = key_desc_part in row_str
            
            if has_price and has_qty and has_desc:
                found_perfect_row = True
                break # Mamy to!
            
            if has_price or has_desc:
                found_fragmented_data = True

        # Krok B: Ocena "≈ömieci" (Garbage Rows)
        # Oczekujemy 1 wiersza danych. Je≈õli jest ich > 3 (np. nag≈Ç√≥wki + dane + ≈õmieci), to b≈ÇƒÖd struktury.
        is_clean_structure = rows_count <= 3 
        
        # Krok C: Finalny Werdykt
        if found_perfect_row:
            if is_clean_structure:
                table_status = "‚úÖ Tabela wczytana bezb≈Çƒôdnie"
            else:
                table_status = "‚ö†Ô∏è Tabela zawiera dane, ale du≈ºo ≈õmieci"
                details.append(f"Wykryto a≈º {rows_count} wierszy (oczekiwano ok. 1)")
        elif found_fragmented_data:
            table_status = "‚ö†Ô∏è Dane rozrzucone / B≈Çƒôdy formatowania"
            details.append("Dane sƒÖ, ale nie w jednym wierszu (np. pociƒôty opis lub przesuniƒôte kolumny)")
        else:
            table_status = "‚ùå Tabela zawiera b≈Çƒôdne dane"
            details.append("Nie znaleziono kluczowych warto≈õci ($85.00, 1.00)")

    else:
        details.append("DataFrame jest pusty lub None")

    # Wypisanie wyniku
    details_str = f"({', '.join(details)})" if details else ""
    print(f"Status tabeli: {table_status} {details_str}")
    if extracted_df is not None:
        print(f"Wymiary tabeli: {extracted_df.shape} (Wiersze x Kolumny)")

    # Zapisz do raportu
    final_report[lib_name] = {
        "Invoice #": inv_status,
        "Recipient": rec_status,
        "Table Status": table_status,
        "Details": details_str
    }

print("‚úÖ Setup Strict Mode zako≈Ñczony.")
```
