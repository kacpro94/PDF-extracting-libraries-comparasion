```python
import pypdf
import camelot
from rapidfuzz import fuzz
import time
```

```python
pdf_path = "data/texts/oldmansea-20.pdf" 

print(f"--- Rozpoczynam test pypdf dla: {pdf_path} ---")

start_time = time.perf_counter()

extracted_text = ""
try:
    reader = pypdf.PdfReader(pdf_path)
    for page in reader.pages:
        extracted_text += page.extract_text() + "\n"
        
except Exception as e:
    extracted_text = f"ERROR: {str(e)}"

end_time = time.perf_counter()
duration = end_time - start_time

score = fuzz.ratio(extracted_text.strip(), ground_truth_simple_1.strip())

print(f"\nCzas wykonania: {duration:.4f} sekundy")
print(f"Dokładność (Similarity Score): {score:.2f}%")
print("-" * 30)
print("PODGLĄD EKSTRAKCJI (Pierwsze 300 znaków):")
print(extracted_text[:300] + "...")
```
